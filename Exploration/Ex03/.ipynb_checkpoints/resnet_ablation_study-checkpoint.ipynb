```python
def imshow(img):
    img = img / 2 + 0.5    # 정규화 되돌기기
    npimg = img.numpy()    # matplotlib은 NumPy 배열을 쓰기 때문에 변환이 필요함
    return np.transpose(npimg, (1, 2, 0))    # 차원 순서 변경 (C, H, W) -> (H, W, C)
```

```python
def show_multiple_images(dataset, n_images=9):
    dataiter = iter(dataset)
    images, labels = next(dataiter)
    fig, axes = plt.subplots(3, 3, figsize=(6, 6))
    axes = axes.flatten()

    for i in range(n_images):
        ax = axes[i]
        img = imshow(images[i])
        ax.imshow(img)
        ax.set_title(f"Label: {trainset.classes[labels[i]]}")
        ax.axis('off')

    plt.tight_layout()  # 각 subplot 크기, 제목, 축, 그림 전체 크기를 전부 고려해서 서로 겹치지 않도록함
    plt.show()
```

## (2) 블록 구성하기

### VGG란?
- "구조를 단순하게 유지하면서 깊이를 늘리면 성능이 좋아진다"는 것을 처음으로 강하게 보여준 모델
- 큰 필터 대신 3X3 Convolution만 반복해서 사용
    - 파라미터 수↓
    - 비선형성(ReLU)이 많이 들어감
    - 표현력↑
- 네트워크를 매우 깊게 쌓음
- 규칙적인 구조 (Conv(3x3) → ReLU → MaxPool)
- 핵심은 "네트워크를 단순하게 유지한 채로 깊게 쌓으면, 표현력이 폭발적으로 증가한다"
- 이 개념 이후 ResNet, DenseNet 같은 모델로 이어짐

#### VGG 장점
- 구조가 직관적이고 이해하기 쉬움
- CNN 기본 구조 학습에 가장 좋은 모델
- Feature extractor로 여전히 많이 쓰임

#### VGG 단점
- 파라미터 수가 매우 많음
- 메모리 사용량이 큼
- 연산량이 많아 속도가 느림

```python
# function for building VGG Block

def build_vgg_block(input_layer, num_cnn=3, channel=64, block_num=1):

    # 입력 레이어
    x = input_layer

    # CNN 레이어
    layers = []
    in_channels = x.size(1)
    for cnn_num in range(num_cnn):
        layers.append(
            nn.Conv2d(
                in_channels=in_channels,
                out_channels=channel,
                kernel_size=3,
                stride=1,
                padding=1
            )
        )

        layers.append(nn.ReLU(inplace=True))
        in_channels = channel

    # Max Pooling 레이어
    layers.append(
        nn.MaxPool2d(
            kernel_size=2,
            stride=2
        )
    )

    # Sequential으로 레이어 묶기
    block = nn.Sequential(*layers) # PyTorch가 이해하는 nn.Module 객체로 만들어줌
    return block
```

```python
class VGGNet(nn.Module):
    def __init__(self):
        super(VGGNet, self).__init__()

        # VGG 블록 생성
        self.vgg_block = build_vgg_block(torch.zeros(1, 3, 32, 32))

    def forward(self, x):
        return self.vgg_block(x)
```

```python
# 블록 1개짜리 model 생성
model = VGGNet()
print(model)

dummy_input = torch.zeros(1, 3, 32, 32)
output = model(dummy_input)
print(output.shape)
```

## (3) VGG Complete Model

```python
# VGG 모델 자체를 생성하는 클래스
class VGG(nn.Module):
    def __init__(self, num_cnn_list=[2, 2, 3, 3, 3], channel_list=[64, 128, 256, 512, 512], num_classes=10):
        
        # num_cnn_list : 각 블록마다 Conv 몇 개 쓸지 (기본값은 VGG-16 구조와 동일)
        # channel_list : 각 블록의 출력 채널 수
        # num_classes 최종 분류 클래스 수 (CIFAR-10 이면 10)
        
        super(VGG, self).__init__()    # nn.Module 초기화 → 파라미터 관리, GPU 이동, Optimizer 등록 가능
        
        assert len(num_cnn_list) == len(channel_list), "num_cnn_list와 channel_list의 길이가 일치해야 합니다."
        # assert : 조건이 참인지 검사하고, 거짓이면 즉시 에러를 발생시키는 파이썬 내장 키워드
        
        layers = []
        in_channels = 3    # 모델에 처음 들어오는 입력 데이터는 채널이 3개(RGB 이미지)라는 전제를 깔고 시작

        for num_cnn, out_channels in zip(num_cnn_list, channel_list):
            layers.append(self._make_vgg_block(in_channels, out_channels, num_cnn))
            in_channels = out_channels    # 다음 블록의 입력 채널을 설정

        self.feature_extractor = nn.Sequential(*layers)    # 특징 추출기
        self.classifier = nn.Sequential(  # FC
            nn.Flatten(),                 # (N, 512, 1, 1) → (N, 512*1*1)    // N : batch size
            nn.Linear(512 * 1 * 1, 4096), # (4096 X 512) X (512 X 1) = (4096 X 1)
            nn.ReLU(True),                # inplace=True → 새로운 결과를 다른 메모리에 만들지 않고, 기존 텐서 자체를 바로 수정함
                                              # 메모리 절약 + 약간 더 빠른 연산, 단 역전파 과정에서 문제가 생길 수 있음
            nn.Linear(4096, 4096),        # (4096 X 4096) X (4096 X 1) = (4096 X 1)
            nn.ReLU(True),
            nn.Linear(4096, num_classes)  # (10 X 4096) X (4096 X 1) = (10 X 1)
        )

    def _make_vgg_block(self, in_channels, out_channels, num_cnn):
        # "Conv 블록 하나"를 만들어주는 코드
        # VGG 기본 단위 : (Conv → ReLU) 여러 번 + MaxPooling 1번

        # in_channels : 입력 채널 수
        # out_channels : 이 블록이 만들어낼 채널 수
        # num_cnn : 이 블록 안에 Conv를 몇 개 쌓을지
        
        layers = [
            nn.Conv2d(in_channels, out_channels, kernel_size=3, padding=1),
            nn.ReLU(True)
        ]

        for _ in range(num_cnn - 1):
            layers.append(nn.Conv2d(out_channels, out_channels, kernel_size=3, padding=1))
            layers.append(nn.ReLU(True))

        layers.append(nn.MaxPool2d(kernel_size=2, stride=2))    # Max Pooling 추가

        return nn.Sequential(*layers)


    def forward(self, x):    # x : 모델에 들어오는 입력 데이터
        x = self.feature_extractor(x)
        x = self.classifier(x)

        return x    # (N, 10)
```

```python
# 기본값을 그대로 사용해서 VGG 모델을 만들면 VGG-16이 됩니다.

vgg_16 = VGG()
summary(vgg_16)
```

```python
# VGG-19 모델 만들기
# 원하는 블록의 설계에 따라 매개변수로 리스트를 전달해 줍니다.
vgg_19 = VGG(
    num_cnn_list=[2, 2, 4, 4, 4],
    channel_list=[64, 128, 256, 512, 512]
)
summary(vgg_19)
```

## (4) VGG-16 vs VGG-19

```python
BATCH_SIZE = 256
EPOCH = 15
```

```python
# CIFAR-10 데이터셋에 대해 Normalize와 Tensor 변환을 적용하는 코드
transform = transforms.Compose([
    transforms.ToTensor(),  # 이미지를 Tensor로 변환
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])  # VGG-16 표준 정규화
])

trainset = torchvision.datasets.CIFAR10(root='./data', train=True, download=True, transform=transform)
testset = torchvision.datasets.CIFAR10(root='./data', train=False, download=True, transform=transform)

trainloader = DataLoader(trainset, batch_size=BATCH_SIZE, shuffle=True)
testloader = DataLoader(testset, batch_size=BATCH_SIZE, shuffle=False)
```

```python
# VGG-16 모델을 만들고 훈련

import time

current_time = time.time()

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")    # GPU를 쓸지, CPU를 쓸지
vgg_16 = torchvision.models.vgg16(pretrained=True)    # 이미 학습된 VGG-16 모델을 불러오는 코드
# pretrained=True : ImageNet 데이터셋으로 이미 학습이 끝난 가중치를 같이 불러온다는 뜻
vgg_16.to(device)    # 모델을 연산 장치로 보내는 코드


for param in vgg_16.parameters():    # 여기서 param은 "숫자 하나"가 아니라 하나의 파라미터 텐서
    param.requires_grad = True  # VGG-16에 들어 있는 모든 파라미터를 학습 가능 상태로 설정
    
criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(    # SGD : 확률적 경사 하강법
    vgg_16.parameters(),  # vgg_16.parameters() : 학습할 모든 파라미터
    lr=0.001,             # lr = 0.001 (learning rate)
    momentum=0.9,         # momentum = 0.9 : 이전 업데이트 방향을 일부 유지해서 더 빠르고 부드럽게 최적화
    weight_decay=1e-4     # weight_decay = 1e-4 : L2 정규화, 가중치가 너무 커지는 걸 방지 (과적합 억제)
)  

# 학습 손실과 검증 정확도를 저장해서 나중에 그래프로 그릴 수 있음
vgg_16_train_losses = []
vgg_16_val_accuracy = []

# Epoch 루프 시작
for epoch in range(EPOCH):
    vgg_16.train()      # 모델 객체 내부의 변수를 vgg_16.training 변수를 True로 변경
                        # training 상태로 바뀌면 Dropout, BatchNorm 사용 규칙 변경됨
    running_loss = 0.0  # loss 누적용 변수(0.0을 씀으로써 자료형을 실수(float)로 고정)

    # 이 두 줄은 훈련 정확도(accuracy)를 계산하기 위한 카운터 초기화
    correct = 0    # 지금까지 본 전체 데이터 개수
    total = 0      # 그중에서 모델이 맞힌 개수

    for i, (inputs, labels) in enumerate(trainloader, 0):
        # trainloader : 배치 단위로 데이터를 가져옴
        # inputs : 이미지 데이터
        # labels : 정답 클래스
        
        inputs, labels = inputs.to(device), labels.to(device)
        # CPU → GPU 또는 GPU → CPU로 데이터 이동
        # 모델이 있는 장치와 데이터를 같은 장치로 맞춰야 연산 가능
        
        optimizer.zero_grad()
        # 이전 배치에서 계산된 gradient를 모두 0으로 초기화
        # PyTorch는 gradient를 누적시키기 때문에 반드시 필요함
        
        outputs = vgg_16(inputs)  # 입력 이미지에 대한 최종 결과값들
        loss = criterion(outputs, labels)  # loss(얼마나 틀렸는지)를 계산함
        loss.backward()  # 이 가중치를 얼마나, 어느 방향으로 바꿔야하는지를 Gradient로 계산
        optimizer.step()  # 앞서 계산된 gradient를 이용해서 가중치를 실제로 수정함

        running_loss += loss.item()  # 현재 batch에서 나온 loss값을 누적 → epoch 전체평균 loss를 계산하기 위함
        _, predicted = torch.max(outputs, 1)  # 각 이미지마다 가장 큰 값을 가진 클래스 인덱스를 뽑음
        total += labels.size(0)  # 이번 batch에 들어온 샘플 수를 전체 개수에 더함
        correct += (predicted == labels).sum().item()  # 이번 배치에서 맞춘 문제 수를 누적함

        if i % 100 == 99:
            print(f"[{epoch + 1}, {i + 1:5d}] loss: {running_loss / (i + 1):.3f}") # 코드수정 i+1 → (i+1) 

    train_loss = running_loss / len(trainloader)  # running_loss: 이번 epoch 동안 모든 batch의 loss를 더한 값
    train_acc = 100 * correct / total  # 이번 epoch의 훈련 정확도(%) 계산
    vgg_16_train_losses.append(train_loss)  # epoch별 training loss를 리스트에 저장

    print(f"Epoch {epoch + 1}: Train Accuracy: {train_acc:.2f}%")

    vgg_16.eval()  # 모델을 평가 모드(Evaluation mode)로 바꿈 / Dropout 꺼짐, batch 통계 사용 X
    correct = 0  # 검증 정확도를 새로 계산하기 위해 카운터를 초기화 함
    total = 0

    with torch.no_grad():  # 학습이 아니라 평가이기 때문에 gradient 계산을 완전히 끔 → 메모리 절약, 연산 속도 증가, 실수로 가중치가 바뀌는 것 방지
        for inputs, labels in testloader:  # testloader에서 batch 단위로 데이터를 가져옴
            inputs, labels = inputs.to(device), labels.to(device)  # 모델이 있는 장치(GPU/CPU)로 데이터 이동
            outputs = vgg_16(inputs) # 순전파 / 모델이 예측한 결과값(logits) 생성
            _, predicted = torch.max(outputs, 1)
            total += labels.size(0)  # 첫 번째 차원의 크기, 여기서는 이미지 개수 (batch size)
            correct += (predicted == labels).sum().item()

    val_acc = 100 * correct / total
    vgg_16_val_accuracy.append(val_acc)

    print(f"Epoch {epoch + 1}: Validation Accuracy: {val_acc:.2f}%")

print("Finished Training")
print(time.time() - current_time)


# 이 코드에서는 학습 중간에 성능이 좋아지는지 확인하기 위해
# 그리고 과적합이 시작되는 시점을 찾기 위해
# validation을 매 epoch마다 확인하고 있음
```

```python
# VGG-19 모델
current_time = time.time()

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
vgg_19 = torchvision.models.vgg19(pretrained=True)
vgg_19.to(device)

for param in vgg_19.parameters():
    param.requires_grad = True
criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(vgg_19.parameters(), lr=0.001, momentum=0.9, weight_decay=1e-4)

vgg_19_train_losses = []
vgg_19_val_accuracy = []

for epoch in range(EPOCH):
    vgg_19.train()
    running_loss = 0.0
    correct = 0
    total = 0

    for i, (inputs, labels) in enumerate(trainloader, 0):
        inputs, labels = inputs.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = vgg_19(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()
        _, predicted = torch.max(outputs, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

        if i % 100 == 99:
            print(f"[{epoch + 1}, {i + 1:5d}] loss: {running_loss / i + 1:.3f}")

    train_loss = running_loss / len(trainloader)
    vgg_19_train_losses.append(train_loss)

    train_acc = 100 * correct / total
    print(f"Epoch {epoch + 1}: Train Accuracy: {train_acc:.2f}%")

    vgg_19.eval()
    correct = 0
    total = 0

    with torch.no_grad():
        for inputs, labels in testloader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = vgg_19(inputs)
            _, predicted = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

    val_acc = 100 * correct / total
    vgg_19_val_accuracy.append(val_acc)

    print(f"Epoch {epoch + 1}: Validation Accuracy: {val_acc:.2f}%")

print("Finished Training")
print(time.time() - current_time)
```
